```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn import tree
import pydotplus
import matplotlib.image as pltimg
```

```python
df = pd.read_csv('C:\\Users\\HP\\python\\DecisionTree\\maharashtrian.csv')
```

```python
d = {'Adnaik Sushant':0, 'Gorule Hemant':1, 'Gunde Swagati':2, 'Kadate Prakash':3, 'Kesarkar Aashish':4, 'Kokitkar Komal':5, 'Kolekar Aishwarya':6, 'Mane Dipak':7,'Mohite Gouri':8, 'Mulani Aalisha':9, 'Patil Rutuja':10, 'Sadale Abhishek':11, 'Suryawanshi Shubham':12, 'Suryawanshi Divya':13, 'Kumbhar Shubham':14}
df['Name'] = df['Name'].map(d)
```

```python
features = ['Age', 'Vada Pav', 'Misal', 'Pohe']

X = df[features]
y = df['Maharashtrian']

print(X)
print(y)
```

```python
#Training and testing spliting
X_train, X_test, Y_train, Y_test = train_test_split(X,y,random_state=0)
#making an instance of the model
clf = DecisionTreeClassifier(max_depth=2, random_state = 0)
#train the model on the data
clf.fit(X_train, Y_train)
#predicting labels of test data
clf.predict(X_test)
```

```python
tree.plot_tree(clf);
```

```python
fn = ['Age', 'Vada Pav', 'Misal', 'Pohe']
cn = ['Maharashtrian', 'Non-Maharashtrian']
fig, axes = plt.subplots(nrows = 1, ncols =1, figsize =(4,4), dpi = 300)
tree.plot_tree(clf, feature_names = fn, class_names = cn, filled = True);
fig.savefig('imagenames.png')
```
